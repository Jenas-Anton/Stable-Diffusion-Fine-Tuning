```python
import diffusers
import gradio as gr
from PIL import Image
from torch import autocast
from diffusers import StableDiffusionPipeline
import torch


# Define the image grid function
def image_grid(imgs, rows, cols):
    assert len(imgs) == rows * cols
    w, h = imgs[0].size
    grid = Image.new('RGB', size=(cols * w, rows * h))

    for i, img in enumerate(imgs):
        grid.paste(img, box=(i % cols * w, i // cols * h))
    return grid

# Inference function with negative prompt and improved parameters
def inference(prompt, num_samples, negative_prompt="", num_inference_steps=100, guidance_scale=8.5, image_height=512, image_width=512):
    all_images = []
    images = pipe(prompt,
                  negative_prompt=negative_prompt,
                  num_images=num_samples,
                  num_inference_steps=num_inference_steps,
                  guidance_scale=guidance_scale,
                  height=image_height,
                  width=image_width).images
    all_images.extend(images)
    return all_images

# Gradio interface
with gr.Blocks() as demo:
    gr.HTML("<h2 style='font-size: 2em; font-weight: bold' align='center'>Stable Diffusion Dreambooth</h2>")

    with gr.Row():
        with gr.Column():
            prompt = gr.Textbox(label="Prompt")
            negative_prompt = gr.Textbox(label="Negative Prompt", placeholder="Describe elements to avoid")
            samples = gr.Slider(label="Samples", value=1, minimum=1, maximum=4, step=1)
            num_inference_steps = gr.Slider(label="Inference Steps", value=100, minimum=10, maximum=200, step=10)
            guidance_scale = gr.Slider(label="Guidance Scale", value=8.5, minimum=1.0, maximum=20.0, step=0.5)
            image_height = gr.Slider(label="Image Height", value=512, minimum=256, maximum=1024, step=64)
            image_width = gr.Slider(label="Image Width", value=512, minimum=256, maximum=1024, step=64)
            run = gr.Button(value="Run")

        with gr.Column():
            gallery = gr.Gallery(show_label=False)

    run.click(inference, inputs=[prompt, samples, negative_prompt, num_inference_steps, guidance_scale, image_height, image_width], outputs=gallery)

    gr.Examples(
        [["A high-quality photo of an SKS watch on a man's wrist", 1, "blurry, distorted, extra limbs", 100, 8.5, 512, 512]],
        [prompt, samples, negative_prompt, num_inference_steps, guidance_scale, image_height, image_width],
        gallery, inference, cache_examples=False
    )

demo.launch(debug=True)
```
